```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
entities = json.load(open('result.json'))
```

```python
fig_consumed, ax_consumed = plt.subplots()
fig_produced, ax_produced = plt.subplots()


for entity in entities:
    if 'Grid' in entity.keys():
        name = entity['Grid']['name']
        produced = entity['Grid']['produced']
        consumed = entity['Grid']['consumed']

        ax_consumed.plot(consumed, label=f'{name} consumed')
        ax_produced.plot(produced, label=f'{name} produced')

    


    if 'Consumer' in entity.keys():
        name = entity['Consumer']['name']
        consumed = entity['Consumer']['consumed']

        ax_consumed.plot(consumed, label=f'{name} consumed')


    if 'Producer' in entity.keys():

        name = entity['Producer']['name']
        produced = entity['Producer']['produced']

        ax_produced.plot(produced, label=f'{name} produced')


ax_consumed.legend()
ax_produced.legend()


```
